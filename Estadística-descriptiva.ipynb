# Variables Categóricas - Categorical Variables

```python
df["Lluvia_SN"] = np.where(df['Lluvia'] > 0, 'Si','No')
# Lluvia_SN <- ifelse(dfMedidas$Lluvia > 0, "Si", "No") #Si las precipitaciones son mayor que 0->Si
# Lluvia_SN <- ifelse(dfMedidas$Lluvia > 0, "Si", "No") #If rain is greater than 0-> Yes
```

```python
df["PM2_half_UMBRAL"] = np.where(df['PM2.5'] >= 25, 'Si','No')
# PM2.5_UMBRAL <- as.factor(ifelse(dfMedidas$PM2.5 >= 25, "Si", "No")) #Si PM2.5 supera umbral ->Si
# PM2.5_UMBRAL <- as.factor(ifelse(dfMedidas$PM2.5 >= 25, "Si", "No")) #If PM2.5 is exceed threshold-> Yes
```

```python
df["PM10_UMBRAL"] = np.where(df['PM10'] >= 50, 'Si','No')
# PM10_UMBRAL  <- as.factor(ifelse(dfMedidas$PM10 >= 50, "Si", "No")) #Si PM10 supera umbral ->Si
# PM10_UMBRAL  <- as.factor(ifelse(dfMedidas$PM10 >= 50, "Si", "No")) #If PM1o is exceed threshold-> Yes
```

```python
df["T_MAX_ALERTA"] = np.where(df['T_MAX'] >= 36, 'A','V')
```

```python
df["T_MAX_ALERTA"] = np.where(df['T_MAX'] >= 39, 'N', df["T_MAX_ALERTA"])
```

```python
df["T_MAX_ALERTA"] = np.where(df['T_MAX'] >= 42, 'R', df["T_MAX_ALERTA"])
# T_MAX_ALERTA  <- as.factor(ifelse(T_MAX >= 42, "R", ifelse(T_MAX >=39, "N", ifelse(T_MAX >=36, "A", "V"))))
```
